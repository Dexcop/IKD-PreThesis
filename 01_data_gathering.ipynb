```python
from google_play_scraper import Sort, reviews_all
import pandas as pd 
import time 
```

```python
APP_ID = 'gov.dukcapil.mobile_id'
LANG = 'id'
COUNTRY = 'id'
```

```python
print(f'Beginning to Scrape the Review for {APP_ID}')
start_time = time.time()

result = reviews_all(
    APP_ID,
    sleep_milliseconds=0,
    lang=LANG,
    country=COUNTRY,
    sort=Sort.NEWEST
)

end_time = time.time()
print(f'Done, Finished getting {len(result)} in {end_time - start_time:.2f} second')
```

```python
df = pd.DataFrame(result)
```

```python
df.head()
```

```python
df_clean = df[['content', 'score', 'at', 'reviewCreatedVersion']]
df_clean.columns = ['review_text', 'rating', 'date', 'app_version']
```

```python
df_clean.head()
```

```python
print(f'Data size: {df_clean.shape}')
print(f'\nSum of null: \n{df_clean.isnull().sum()}')
print(f"\nRating distribution: \n{df_clean['rating'].value_counts().sort_index()}")
```

```python
filename = 'review_ikd_raw.csv'
filepath = f'./Result/{filename}'

df_clean.to_csv(filepath, index=False)

print(f'Data saved to {filepath}')
df_clean.head()
```
